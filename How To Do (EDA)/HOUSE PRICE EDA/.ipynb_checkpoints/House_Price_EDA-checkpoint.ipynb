```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
df=pd.read_csv('train.csv')
```

```python
print(df.columns.values)
```

```python
df['SalePrice'].describe()
```

```python
sns.distplot(df['SalePrice'])
plt.show()
```

```python
var = 'GrLivArea'
data = pd.concat([df['SalePrice'], df[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice')
```

```python
var = 'TotalBsmtSF'
data = pd.concat([df['SalePrice'], df[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice')
```

```python
var = 'OverallQual'
data = pd.concat([df['SalePrice'], df[var]], axis=1)
plt.subplots(figsize=(8,6))
fig = sns.boxplot(x=var, y="SalePrice", data=df)
```

```python
var = 'YearBuilt'
data = pd.concat([df['SalePrice'], df[var]], axis=1)
plt.subplots(figsize=(16, 8))
fig = sns.boxplot(x=var, y="SalePrice", data=df)
plt.xticks(rotation=90);
```

```python
#correlation matrix
corrmat = df.select_dtypes(include=['number']).corr()
plt.figure(figsize=(6,4))
sns.heatmap(corrmat, vmax=.8, square=True,);
```

```python
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df[cols].values.T)
plt.figure(figsize=(4,4))
sns.set(font_scale=1)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 9.5}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
#scatterplot
sns.set()
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']
sns.pairplot(df[cols], size = 2.5)
plt.show();
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

# Cleaning Data (Steps)
1. Always Make a Copy of Your Dataset Before Cleaning Data
2. Messy Data
3. Dirty Data

```python
students_copy=students.copy()
```

```python
# Removing 'AI_Tools_Used' col permanent
students_copy.drop(columns={'AI_Tools_Used'},inplace=True)
```

```python
students_copy.columns
```

```python
# replacing 1,2,3,4,5 in string value like 1.Very low Trust, 2.low Trust, 3.Moderate Trust, 4. Strong Trust, 5.Very High Trust permanently
students_copy['Trust_in_AI_Tools']=students_copy['Trust_in_AI_Tools'].replace({1:'Very low Trust',2:'Low Trust',3:'Moderate Trust',4:'Strong Trust',5:'Very High Trust'})
```

```python
# replacing -1,-2,-3,0,+1,+2,+3 in string value like -1. Slightly negative impact, -2. Moderate Negative Impact, -3. Strong Negative Impact, 0. No Impact, +1. SLightly Positive Impact, +2. Moderate Positive Impact , +3 Strong Positive Impact
students_copy['Impact_on_Grades']=students_copy['Impact_on_Grades'].replace({+1:'SLightly Positive Impact',+2:'Moderate Positive Impact',+3:'Strong Positive Impact',0:'No Impact',-1:'Slightly Negative Impact',-2:'Moderate Negative Impact',-3:'Strong Negative Impact'})
```

```python
# Replacing All the value from Awareness_Level column
# Define the mapping
awareness_map = {
    1: 'No Awareness',
    2: 'Minimal Awareness',
    3: 'Very Basic Awareness',
    4: 'Basic Awareness',
    5: 'Fair Awareness',
    6: 'Moderate Awareness',
    7: 'Good Awareness',
    8: 'Strong Awareness',
    9: 'Very High Awareness',
    10: 'Expert-Level Awareness'
}
# Apply the mapping
students_copy['Awareness_Level'] = students_copy['Awareness_Level'].replace(awareness_map)
```

```python
# Replacing "_" to "space" in columns names
students_copy.columns=students_copy.columns.str.replace('_',' ')
```

```python
students_copy.rename(columns={'Internet Access':'Internet Speed'},inplace=True)
```

```python
students_copy.columns
```

```python
# droping all students which does not have state name keeping only those rows who has state name and all information
students_copy = students_copy[students_copy['State'].notnull()]
```

# Cleaning

```python
df[['Surname','Name']]=df['Name'].str.split(',',n=1,expand=True)
```

```python
df['Names']=df['Name']+' '+df['Surname']
```

```python
df.drop(columns={'Name','Surname'},inplace=True)
```

```python
df.shape[0]
```

```python
# The peoples which does not have age 
df[df['Age'].isna()]
```

```python
# The people which has age values
df=df[df['Age'].notna()]
```

```python
df
```

```python
df['TicketNumber'] = df['Ticket'].str.extract(r'(\d+)$')
```

```python
df.drop(columns={'Ticket'},inplace=True)
```

```python
# Cabin col has more than 50% missing Values so it is droped From dataset
df.drop(columns={'Cabin'},inplace=True)
```

```python
df=df[df['Embarked'].notna()]
```

```python
df=df[df['TicketNumber'].notna()]
```

```python
df['Age']=df['Age'].astype(int)
```

```python
df.rename(columns={'TicketNumber':'Ticket Number'},inplace=True)
```

```python
df['Ticket Number']=df['Ticket Number'].astype(int)
```

```python
df.head(10)
```

```python
# Memory Size has also been reduced from 83.7 to 66.4 kb
df.info()
```

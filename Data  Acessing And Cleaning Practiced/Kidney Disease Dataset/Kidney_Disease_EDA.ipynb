# Information About Normal Levels :-

1. Creatinine_Level column :- (0.74-1.35) normal
                              (>1.35) kidney diseases(eg.stones)
                              (<0.74) muscle loss.
   
3. Blood_Urea_Level column :- (<7) overdehydration,liver disease
                               (7-20) Healthy kidney,Liver Function
                               (>20) Kidney Disease
   
4. Glomerular_Filteration_Rate column :- (>90) (Normal Stage(1))
                                         (60-89) (Stage(2) normal in case of adults)
                                         (45-59) (Stage(3a) early sign of damage)
                                         (30-44) (Stage(3b) Worsening Function)
                                         (15-29) (Stage(4) Lossing Function need of dialysis or transplant)
                                         (<15)   (Stage(5) Kidney Loss all Function need transplant or dialysis)
5. Urine_Level :- (800-2000ml/day) normal
                  (<400) oliguria
                  (<100) Anuria
                  (>2500-3000) Polyuria

# Before doing Multivariate Analysis we have to convert the numerical columns values in categorical values because numerical values/levels are difficult to understand so to make it easier we will do feature enginerring.

```python
def creatine_level(level):
    if level < 0.59:
        return 'Low'
    elif 0.59 <= level <= 1.35:
        return 'Normal'
    else:
        return 'High'
```

```python
df['Creatine_Levels']=df['Creatinine_Level'].apply(creatine_level)
```

```python
df.drop(columns={'Creatinine_Level'},inplace=True)
```

```python
df.head()
```

```python
df['Creatine_Levels'].value_counts()
```

```python
def categorize_blood_urea(urea):
    if urea < 7:
        return 'Low'
    elif 7 <= urea <= 20:
        return 'Normal'
    else:
        return 'High'
```

```python
df['Blood_Urea_Levels']=df['Blood_Urea_Level'].apply(categorize_blood_urea)
```

```python
df.drop(columns={'Blood_Urea_Level'},inplace=True)
```

```python
df['Blood_Urea_Levels'].value_counts()
```

```python
def gfr_stage_simple(gfr):
    if gfr >= 90:
        return '(1) Normal'
    elif 60 <= gfr < 90:
        return '(2) Mildly Decreased'
    elif 30 <= gfr < 60:
        return '(3) Moderate'
    elif 15 <= gfr < 30:
        return '(4) Severe'
    else:
        return '(5) Kidney Failure'
```

```python
df['GFR_Stage'] = df['Glomerular_Filteration_Rate'].apply(gfr_stage_simple)  # or gfr_stage_simple
```

```python
df[['Glomerular_Filteration_Rate','GFR_Stage']].head(10)
```

```python
df.drop(columns={'Glomerular_Filteration_Rate'},inplace=True)
```

```python
df.head()
```

```python
def categorize_urine_output(urine_ml):
    if pd.isnull(urine_ml):
        return 'Unknown'
    elif urine_ml < 100:
        return '(5) Anuria'
    elif 100 <= urine_ml < 400:
        return '(4) Oliguria'
    elif 400 <= urine_ml < 800:
        return '(3) Low'
    elif 800 <= urine_ml <= 2000:
        return '(2) Normal'
    else:
        return '(1) Polyuria'
```

```python
df['Urine_Levels'] = df['Urine_Level'].apply(categorize_urine_output)
```

```python
df[['Urine_Level','Urine_Levels']].head(10)
```

```python
df.drop(columns={'Urine_Level'},inplace=True)
```
